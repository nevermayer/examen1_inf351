```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree

# Cargar el dataset Iris.
data = pd.read_csv("IRIS.csv")

# Dividir el dataset en conjunto de entrenamiento y conjunto de prueba.
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.2)

# Construir el √°rbol de decisi√≥n.
arbol = DecisionTreeClassifier(max_depth=3)
arbol.fit(X_train, y_train)

# Predecir las etiquetas del conjunto de prueba.
y_pred = arbol.predict(X_test)

# Calcular la precisi√≥n del modelo.
precision = accuracy_score(y_test, y_pred)

# Imprimir la precisi√≥n del modelo.
print("Precisi√≥n:", precision)

# Visualize the decision tree
plt.figure(figsize=(20, 20))
plt.title("√Årbol de decisi√≥n")
plt.xlabel("Caracter√≠sticas")
plt.ylabel("Etiquetas")
tree.plot_tree(arbol, filled=True)
plt.show()
```

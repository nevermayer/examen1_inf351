```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Leer el archivo CSV

df = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
# Imprimir las columnas disponibles para que el usuario las vea
print("Columnas disponibles:")
for i, columna in enumerate(df.columns):
    print(f"{i + 1}. {columna}")

# Solicitar al usuario que ingrese los números de las columnas de interés separados por comas
columnas_seleccionadas = '3,4,5'
columnas_seleccionadas = [int(idx) - 1 for idx in columnas_seleccionadas.split(",")]
```

```python
# Crear subplots para los gráficos
fig, axs = plt.subplots(len(columnas_seleccionadas), figsize=(10, 10 * len(columnas_seleccionadas)))

# Calcular la media, moda, cuartiles y percentiles para las columnas seleccionadas y agregar gráficos
for i, columna_idx in enumerate(columnas_seleccionadas):
    columna = df.columns[columna_idx]
    datos_columna = df[columna]
    
    # Media utilizando numpy
    media = np.mean(datos_columna)
    
    # Moda utilizando pandas
    moda = datos_columna.mode().values[0]  # Puede haber múltiples modas
    
    # Cuartiles utilizando numpy
    cuartiles = np.percentile(datos_columna, [25, 50, 75])
    
    # Percentiles utilizando numpy
    percentil_25 = np.percentile(datos_columna, 25)
    percentil_75 = np.percentile(datos_columna, 75)
    
    # Imprimir resultados
    print(f"Columna: {columna}")
    print(f"Media: {media}")
    print(f"Moda: {moda}")
    print(f"Cuartiles (Q1, Q2, Q3): {cuartiles[0]}, {cuartiles[1]}, {cuartiles[2]}")
    print(f"Percentiles (P25, P75): {percentil_25}, {percentil_75}")
    print("\n")

    # Crear un gráfico de barras para las estadísticas
    axs[i].bar(['Media', 'Moda', 'Q1', 'Q2', 'Q3', 'P25', 'P75'], [media, moda, cuartiles[0], cuartiles[1], cuartiles[2], percentil_25, percentil_75])
    axs[i].set_title(f'Estadísticas de la Columna: {columna}')

# Ajustar la disposición de los gráficos
plt.tight_layout()
plt.show()
```
